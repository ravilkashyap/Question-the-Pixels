```python
import json
from tqdm import tqdm 
import pandas as pd
from collections import Counter
import numpy as np
```

```python
from tqdm import tqdm
```

```python
# Insert image_id, question_id question and answer into a csv


ques_ans=[]

for i,j in tqdm(zip(js['questions'],json_data['annotations'])):
    img_id=i['image_id']
    question=i['question']
    ques_id=i['question_id']
    assert j['question_id']==ques_id , "Question ID mismatch"
    one_ans=[ans['answer'] for ans in j['answers']][0]
    ques_ans.append([img_id,ques_id,question.encode('utf-8'),one_ans.encode('utf-8')]) # encode('utf-8') because str() fails for unicode                 
```

```python
out_df=pd.DataFrame(ques_ans,columns=['image_id','ques_id','question','answer'])
```

```python
out_df.to_csv('train_data.csv',index=False)
```

## Select only top n (n=1000) classes

```python
df = pd.read_csv("../data/train_data.csv")
df.head()
```

```python
answers = []
for _,row in tqdm(df.iterrows()):
    answers.append(str(row.answer))
```

```python
print(len(answers))
answers[:10]
```

```python
len(np.unique(answers))
```

```python
freq = Counter(answers).most_common(1000)
freq
```

```python
top_1000_ans=[]
sum = 0
for val,c in freq:
    sum+=c
    top_1000_ans.append(val)

print(round(sum*100/len(answers),2) ,"%")
```

There are 36255 classes in total, we can't proceed with those many class classification hence we only consider top 1000 classes as it covers about 83% of the entire data

```python
new_data = []
c=0
for _,row in tqdm(df.iterrows()):
    if row.answer in top_1000_ans:
        new_data.append([row.image_id,row.ques_id,row.question,row.answer])
    else:
        c+=1
    
        
```

```python
assert len(new_data)==sum, "Few entries missing check again" #to check if we got the required entries
```
