```python
# Imports

import keras
from keras.preprocessing import image
from keras.applications.densenet import DenseNet121
from keras.applications.densenet import preprocess_input
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
from tqdm import tqdm
```

## Get paths for images

```python
# Get paths for training images

df = pd.read_csv("../data/train_data_1000.csv")
df.head()
```

```python
img_id = df['image_id']
```

```python
img_paths = ['../data/train/images/COCO_train2014_'+'0'*(12-len(str(i)))+str(i)+'.jpg' for i in img_id]
```

```python
assert len(img_id) == len(img_paths)
```

```python
img = image.load_img(img_paths[111111], target_size=(224, 224))
img
```

```python
path_data=[]
for i,j in tqdm(zip(img_id,img_paths)):
    path_data.append([i,j,'0'])
```

```python
out_df=pd.DataFrame(path_data,columns=['image_id','img_path','dummy_class'])
```

```python
%%time
out_df.to_csv('../data/path_train_1000.csv',index=False)
```

```python
out_df.head()
```

```python
test_datagen = ImageDataGenerator(rescale=1./255, preprocessing_function=preprocess_input)
```

```python
test_generator = test_datagen.flow_from_dataframe(
    out_df,
    x_col='img_path', 
    y_col='dummy_class',
    target_size=(224, 224),
    color_mode="rgb",
    shuffle = False,
    drop_duplicates=False,
    batch_size=512)
```
